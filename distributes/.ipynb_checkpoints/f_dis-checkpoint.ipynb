```python
# -*- coding: utf-8 -*-
'''
FX~x^2(m), Y~x^2(n)F=(X/m)/(Y/n)
by: Justin Peng
'''

import numpy as np
import math
import matplotlib.pyplot as plt

m_list = [10]
n_list = [4,10,50,500]
#n
x_length = 10000
#
_scale = 20
#
f_list = []
#
x_data = []   
#
y_data = []
#0~1
plt.figure(figsize=(16,9))
plt.title("F distribute, X scale = %s"%_scale)
plt.xlabel("Random Variable of F=(X/m)/(Y/n)")
plt.ylabel("Probability")

def f_value(_m,_n):
    _mx2 = 0
    _nx2 = 0
    for _p in np.random.randn(_m):
        _mx2 +=  pow(_p,2)
    for _p in np.random.randn(_n):
        _nx2 +=  pow(_p,2)
    return (_mx2/_m)/(_nx2/_n)

for _m in m_list: 
    for _n in n_list:
        f_list.clear()
        #f
        _f = 0
        for num in range(1,x_length+1):
            _f = f_value(_m,_n)
            if _f < 4:
                f_list.append(_f*_scale)
        #4_scale
        f_list.sort()
        #
        _min = min(f_list)
        _max = max(f_list)
        _len = len(f_list)
        #
        _start = 0
        #_start_start+1
        _cnt = 0
        x_data.clear()
        y_data.clear()
   
        for _ln in range(_len):
            if f_list[_ln]>=_start and f_list[_ln]<_start+1:
                _cnt += 1
                #_start_start+1
                if _ln == _len-1:
                    #
                    x_data.append(_start+1)
                    y_data.append(_cnt/_len)
                    break
            else:

                if _ln == _len-1:
                    #
                    x_data.append(_start+1)
                    y_data.append(_cnt/_len)
                    _cnt = 1
                    x_data.append(math.ceil(f_list[_ln]))
                    y_data.append(_cnt/_len)
                    break
                else:
                    #floor1
                    x_data.append(_start+1)
                    y_data.append(_cnt/_len)               
                    _start = math.floor(f_list[_ln])
                    #print(_start)
                    _cnt = 1

        #print(x_data)
        #print(y_data)
        regr = np.poly1d(np.polyfit(x_data,y_data,10))
        #
        y_pred = regr(x_data) 
        plt.scatter(x_data,y_data)
        plt.plot(x_data, y_pred,label="m=%s"%(_m)+",n=%s"%(_n))

plt.legend()
plt.show()
```
